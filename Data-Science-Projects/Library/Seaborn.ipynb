# Seaborn library

```python
import seaborn as sns 
import pandas as pd
```

```python
data_set=pd.read_csv('Titanic-Dataset.csv')
data_set
```

```python
sns.boxplot(y='Age', data=data_set)
```

```python
sns.boxplot(x= 'Survived',y='Age', data=data_set)
```

```python
data_set['Age']=data_set['Age'].fillna(0).astype(int)
```

```python
data_set["Age"] = data_set["Age"].replace(0, data_set["Age"].median())
```

```python
sns.violinplot(x='Age' , y='Sex', data=data_set)
```

```python
data_set['Sex'] = data_set['Sex'].map({"male": 0, "female": 1})
```

```python
sns.regplot(x='Age' , y='Sex', data=data_set)
```
