# NER

```python
from transformers import pipeline
ner= pipeline('ner', grouped_entities=True)
ner('My name is Loay, and I work at Alrazi Company in Tiazz.')
```

# Q&A

```python
from transformers import pipeline

qa = pipeline('question-answering')
qa(
    question = "Where do I work?",
    context = 'My name is Loay, and I work at Alrazi Company in Tiazz.'
)
    


```
