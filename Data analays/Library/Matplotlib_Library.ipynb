# Matplotlib Library

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import pandas as pd
import numpy as np
```

# Plot Data

```python
days= [1,2,3,4,5,6,7]
temp_Taiz=[22.4, 21.3, 22, 21, 23.6, 22.7, 24]
temp_Sanaa=[16.8, 18.3, 20, 21.1, 19.5, 16.7, 17]
```

```python
df_Taiz=pd.DataFrame(data={'Day':days, 'Temperatures':temp_Taiz})
```

```python
df_Sanaa=pd.DataFrame(data={'Day':days, 'Temperatures':temp_Sanaa})
```

```python
df_Taiz
```

```python
df_Sanaa
```

# Line Plots

```python
plt.plot(df_Taiz['Day'], df_Taiz['Temperatures'])
plt.show()
```

# scatter Plots

```python
plt.scatter(df_Taiz['Day'], df_Taiz['Temperatures'])
plt.show()
```

# Bar Charts

```python
plt.bar(df_Taiz['Day'], df_Taiz['Temperatures'])
plt.show()
```

# Customizing Plots

```python
plt.plot(df_Taiz['Day'], df_Taiz['Temperatures'],label='Taiz',ls='--', marker='*')
plt.plot(df_Sanaa['Day'], df_Sanaa['Temperatures'], label='Sanaa' , c='r', alpha=0.4 , marker='o')
plt.xlabel('days of the week')
plt.ylabel('Temperatures in cel')
plt.title('Temperatuers Degrees in Taiz')
plt.legend()# it show Taiz blue and Sanaa= orange color
```

# Subpolts

```python
figure , axes = plt.subplots(1,2, figsize=(10,5))
axes[0].plot(df_Taiz['Day'], df_Taiz['Temperatures'],label='Taiz',ls='--', marker='*')
axes[0].set_title('Temperatures Degree in Taiz')


axes[1].plot(df_Sanaa['Day'], df_Sanaa['Temperatures'],label='Taiz',ls='--', marker='*')
axes[1].set_title('Temperatures Degree in Sanaa')
```
