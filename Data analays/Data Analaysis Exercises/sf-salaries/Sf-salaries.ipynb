```python
# الإبقاء فقط على الصفوف التي لا تساوي 'Not Provided'
df = df[df['BasePay'] != 'Not Provided']
```

```python
# حذف أي صف يحتوي على 'Not Provided' في أي عمود
df = df[~(df == 'Not Provided').any(axis=1)]
```
