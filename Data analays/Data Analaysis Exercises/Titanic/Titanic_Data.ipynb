# Titanic Dataset 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
data_set=pd.read_csv('Titanic-Dataset.csv')
```

# to clean data

```python
print(data_set['Age'].dtype)
```

```python
print(data_set['Age'].unique())
```

```python
print(data_set['Age'].isna().sum()) 
```

```python
data_set['Age']=data_set['Age'].fillna(0).astype(int)
```

```python
data_set["Age"] = data_set["Age"].replace(0, data_set["Age"].median())
```

```python
print(data_set['Sex'].isna())
print(data_set['Sex'].unique())
```

```python
data_set['Sex'] = data_set['Sex'].map({"male": 0, "female": 1})
```

# Analys

```python
# فلترة الصفوف بحيث الجنس female والنجاة Survived = 1
num_survived_women = data_set[(data_set["Sex"] == 1) & (data_set["Survived"] == 1)].shape[0]

print("عدد النساء اللواتي نجين:", num_survived_women)
```

```python
df_Titanic_Data = pd.DataFrame(data_set)
df_Titanic_Data
```

# البقاء مقابل فئة التذكرة

```python
plt.bar(df_Titanic_Data['Survived'], df_Titanic_Data['Pclass'])
plt.xlabel('Survived')
plt.ylabel('Pclass')
plt.title('Pclass versus survival')
```
