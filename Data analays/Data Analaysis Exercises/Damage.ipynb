```python
class SmartAnalyzer:
    def __init__(self, messages):
        self.messages = messages
        # استخدام Dictionary للسرعة القصوى O(1)
        self.stats = {} 

    def analyze_lengths(self):
        for msg in self.messages:
            sender = msg.get('from')
            text = msg.get('text')
            
            # التأكد أن الرسالة نصية وليست صورة أو ملف
            if sender and isinstance(text, str):
                if sender not in self.stats:
                    self.stats[sender] = {'total_words': 0, 'msg_count': 0}
                
                # حساب عدد الكلمات في الرسالة
                word_count = len(text.split())
                
                # تحديث البيانات داخل القاموس
                self.stats[sender]['total_words'] += word_count
                self.stats[sender]['msg_count'] += 1

    def get_results(self):
        print(f"{'الاسم':<20} | {'متوسط طول الرسالة (كلمة)':<25}")
        print("-" * 50)
        for user, data in self.stats.items():
            avg = data['total_words'] / data['msg_count']
            print(f"{user:<20} | {avg:<25.2f}")

# تشغيل المحلل
analyzer = SmartAnalyzer(all_messages) # all_messages هي القائمة المدمجة
analyzer.analyze_lengths()
analyzer.get_results()
```
