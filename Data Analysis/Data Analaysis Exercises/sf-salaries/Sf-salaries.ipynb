# How many people have the word Chief in their Job title?(This is pretty trickly)

```python
def chief (string):
    if 'chief' in (string.lower()):
        return True
    else:
        return False
```

```python
(df['JobTitle'].apply(lambda x : chief(x))).sum()
```
