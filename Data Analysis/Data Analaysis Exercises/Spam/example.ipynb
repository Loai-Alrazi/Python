```python
import joblib

# تحميل الملفات المحفوظة
loaded_model = joblib.load('spam_detector_model.pkl')
loaded_vectorizer = joblib.load('tfidf_vectorizer.pkl')

# تجربة رسالة جديدة فوراً
new_msg = ["You have won a lottery, call us now!"]
new_msg_tfidf = loaded_vectorizer.transform(new_msg)
result = loaded_model.predict(new_msg_tfidf)

print("Spam" if result[0] == 1 else "Normal")
```
